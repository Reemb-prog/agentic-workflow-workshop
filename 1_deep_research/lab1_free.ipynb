### One-time setup

Install the Agents SDK with AnyLLM support:

```bash
pip install "openai-agents[any-llm]" python-dotenv
```

Create an OpenRouter API key and save it in your `.env` file:

```text
OPENROUTER_API_KEY=your_openrouter_api_key
```

This lab uses OpenRouter's free-model router, so it does not call or bill the OpenAI API.


````python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace, set_tracing_disabled
````

````python
# Environment variables are stored in the .env file

load_dotenv(override=True)

# No OpenAI API key is used, so do not upload traces to OpenAI
set_tracing_disabled(disabled=True)
````

````python

# Make an agent with name, instructions, model

agent = Agent(
    name="Jokester",
    instructions="You are a joke teller",
    model="any-llm/openrouter/openrouter/free",
)
````

````python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
````

## Trace behavior

OpenAI hosted tracing is disabled because this lab does not use an OpenAI API key.

The agent request is routed through AnyLLM to OpenRouter's free-model router. Free-model availability and usage limits are controlled by OpenRouter.
